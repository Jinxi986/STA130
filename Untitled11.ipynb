The Simple Linear Regression (SLR) model is a elementary statistical model that demonstrate the relationship between two variables: a predictor variable (independent variable) and an outcome variable (dependent variable). 

Predictor Variable (X): This is the variable which can effects the outcome. 

Outcome Variable (Y): This is the variable which has a linear association with the predictor variable X. 

Slope (Œ≤‚ÇÅ): It represents a shifting degree the outcome variable will generate with the variation of one-unit predictor variable. Also, it reveals the positive and negative correlation between two variable. A positive slope indicates that as the predictor raises, the outcome tends to increaseÔºåand vice versa.

Intercept coefficients(Œ≤‚ÇÄ): This is the value of the outcome variable when the predictor variable is zero. In other words, it sets the baseline degree of the independent variable in the model.

Error Term (Œµ): real data isn't perfectly predictable, thus the error term tells Y's variability that cannot be predicted by ùëã. It shows unknown influences in true world on ùëå that the model doesn't considered.

the Simple Linear Regression model is represented as:

ùëå = ùõΩ0 + ùõΩ1ùëã + œµ

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import statsmodels.formula.api as smf

# Parameters
np.random.seed(0)
n_samples = 100
beta_0 = 2.0       # Intercept
beta_1 = 0.5       # Slope
sigma = 1.0        # Standard deviation of error term

# Simulate predictor variable X
X = np.random.uniform(0, 20, n_samples)

# Calculate outcome variable Y
Y = beta_0 + beta_1 * X + np.random.normal(0, sigma, n_samples)

# Store in a pandas DataFrame
data = pd.DataFrame({'X': X, 'Y': Y})

# Define and fit the model
model = smf.ols('Y ~ X', data=data).fit()

# Print the summary of the model
print(model.summary())

# Plotting the data points
plt.scatter(data['X'], data['Y'], label="Observed Data", color="blue", alpha=0.6)

# Plotting the fitted line from the model
plt.plot(data['X'], model.predict(data), color="red", label="Fitted Regression Line")

# Adding labels and legend
plt.xlabel("Predictor Variable X")
plt.ylabel("Outcome Variable Y")
plt.title("Fitted Simple Linear Regression Model")
plt.legend()
plt.show()

fig.show(renderer="png")
```

In the "ordinary least squares" (OLS) method, the line chosen for the fitted model is the one that minimizes the sum of the squared differences between the observed data points and the predicted values from the model. These differences are referred to as "residuals," and the method seeks to minimize the total residual error by adjusting the model's parameters.

The reason it requires "squares" is to ensure that positive and negative residuals do not cancel each other out. By squaring the residuals, we give more weight to larger errors, which helps in finding the best-fitting line that balances all the data points as closely as possible. This approach also ensures that the solution is mathematically tractable and leads to a unique, optimal fit.

In a simple linear regression model, the null hypothesis of "no linear association (on average)" between two variables is expressed in terms of the slope parameter (ùõΩ1) of the model. For the Old Faithful Geyser dataset, this null hypothesis can be specified as √ü1 = 0. This implies that, on average, there is no linear relationship between the waiting time (predictor) and the duration of eruptions (response).

To test this hypothesis, we can fit a simple linear regression model with waiting time as the predictor for duration and examine the resulting summary, which includes a hypothesis test on the slope (ùõΩ1).

To interpret the beliefs regarding the Old Faithful Geyser dataset after running the linear regression, we need to focus on the p-value for the slope coefficient (ùõΩ1) and its confidence interval.

Coefficient Estimate for Waiting Time (slope ùõΩ1): This is the estimated average change in eruption duration for each additional unit of waiting time. If the p-value associated with this estimate is less than a typical significance level (e.g., 0.05), we would reject the null hypothesis.

p-value for ùõΩ1: This tests the null hypothesis that Œ≤1=0. A low p-value (e.g., < 0.05) provides evidence against the null hypothesis, suggesting a statistically significant association between waiting time and duration.

Confidence Interval for ùõΩ1: If the interval does not contain zero, it further suggests that there is an association.

To summary, if the p-value is low and the confidence interval excludes zero, we would strongly believe that there is a linear association between waiting time and eruption duration.
If the p-value is high and the confidence interval includes zero, we would revise our belief to suggest that there is no significant linear relationship between these variables, and any observed association might be due to random variation.

Bootstrap Sampling: We‚Äôll create multiple bootstrap samples from the observed data with n=160 long wait times and calculate the slope coefficient for each sample. This will help us visualize the empirical distribution of the slope under the observed data.

Simulation under the Null Hypothesis: We‚Äôll simulate new duration values under the assumption that there is no association between waiting and duration (i.e., assuming a slope of zero). By comparing this simulated distribution to the bootstrap distribution, we can assess the probability of obtaining the observed slope coefficient.

```python
import numpy as np
import pandas as pd
import statsmodels.formula.api as smf
import plotly.express as px
import matplotlib.pyplot as plt

# Define the long wait data subset
long_wait_limit = 71
long_wait = old_faithful[old_faithful.waiting > long_wait_limit]

# Fit the model to the full long wait data
model = smf.ols('duration ~ waiting', data=long_wait)
fitted_model = model.fit()

# Print the summary of the model fitted to the full data
print(fitted_model.summary().tables[1])

# Bootstrap Sampling to collect sampling distribution of the slope coefficients
n_bootstrap = 1000  # Number of bootstrap samples
bootstrap_slopes = []

for _ in range(n_bootstrap):
    # Sample with replacement from the long wait data
    bootstrap_sample = long_wait.sample(n=len(long_wait), replace=True)
    # Fit the linear regression model to the bootstrap sample
    bootstrap_model = smf.ols('duration ~ waiting', data=bootstrap_sample).fit()
    # Collect the slope coefficient
    bootstrap_slopes.append(bootstrap_model.params['waiting'])

# Visualize the bootstrapped sampling distribution of the slope coefficients
plt.hist(bootstrap_slopes, bins=30, edgecolor='black', alpha=0.7)
plt.title("Bootstrap Sampling Distribution of the Slope Coefficients")
plt.xlabel("Slope (Estimated Coefficient for waiting)")
plt.ylabel("Frequency")
plt.show()

# Simulate samples under the null hypothesis (beta_1 = 0)
n_simulations = 1000
simulated_slopes = []

# Get the observed values of waiting and duration from the long wait data
waiting_values = long_wait['waiting']
duration_values = long_wait['duration']

for _ in range(n_simulations):
    # Simulate the duration based on the null hypothesis (beta_1 = 0)
    simulated_duration = np.random.normal(loc=duration_values.mean(), scale=duration_values.std(), size=len(waiting_values))
    # Create a new simulated DataFrame
    simulated_data = pd.DataFrame({'waiting': waiting_values, 'duration': simulated_duration})
    # Fit the linear regression model to the simulated data
    simulated_model = smf.ols('duration ~ waiting', data=simulated_data).fit()
    # Collect the slope coefficient
    simulated_slopes.append(simulated_model.params['waiting'])

# Visualize the simulated sampling distribution of the slope under the null hypothesis
plt.hist(simulated_slopes, bins=30, edgecolor='black', alpha=0.7)
plt.title("Simulated Sampling Distribution of the Slope (Under Null Hypothesis)")
plt.xlabel("Slope (Estimated Coefficient for waiting)")
plt.ylabel("Frequency")
plt.show()

# Calculate the 95% bootstrap confidence interval for the slope
lower_bound = np.percentile(bootstrap_slopes, 2.5)
upper_bound = np.percentile(bootstrap_slopes, 97.5)
print(f"95% Bootstrap Confidence Interval for the Slope: ({lower_bound:.4f}, {upper_bound:.4f})")

# Check if the observed slope is contained in the 95% CI
observed_slope = fitted_model.params['waiting']
print(f"Observed Slope: {observed_slope:.4f}")
print(f"Is the observed slope within the 95% CI? {'Yes' if lower_bound <= observed_slope <= upper_bound else 'No'}")

# Calculate the p-value for the simulated slopes under the null hypothesis
simulated_p_value = np.mean(np.abs(np.array(simulated_slopes)) >= np.abs(observed_slope))
print(f"Simulated p-value: {simulated_p_value:.4f}")

# Compare with the p-value from the fitted model
p_value_from_model = fitted_model.pvalues['waiting']
print(f"P-value from the fitted model: {p_value_from_model:.4f}")

fig.show(renderer="png")
```

(A) Permutation Test (Shuffling Labels)
How it works:
The null hypothesis assumes no difference between the groups.
Combine all data, shuffle the labels, and split them into two new groups.
Calculate the difference in means for each shuffle and repeat many times (e.g., 10,000 iterations).
Compare the observed difference to the distribution of shuffled differences to calculate the p-value.

(B) Bootstrap Confidence Interval
How it works:
Resample with replacement from each group to create bootstrap samples (e.g., 10,000 samples).
Calculate the difference in means for each bootstrap sample.
Compute the 95% confidence interval from the 2.5th and 97.5th percentiles of the differences.

(a) How These Sampling Approaches Work
Permutation Test: Shuffles the labels to create a distribution of differences under the null hypothesis and compares the observed difference to this distribution.
Bootstrap: Resamples within each group to create a distribution of the difference in means, then calculates a confidence interval for the true difference.

(b) Comparison with Indicator Variable Model
Similarity:
All methods test the null hypothesis of no difference between groups.
Differences:
Permutation/Bootstrap: Non-parametric, do not assume a specific model or distribution.
Indicator Variable Model: Parametric, assumes a specific model (e.g., linear regression) and tests for group differences using coefficients and p-values.
